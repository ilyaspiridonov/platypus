## Разложение информации на биты

Первое, с чем нужно познакомиться, – это с понятием *битов*. Биты придумали в качестве простейшего алфавита в мире. Любую информацию можно представить всего двумя символами: 0 и 1.

Одним из примеров являются числа. Вы, вероятно, привыкли представлять число с помощью [строки](gloss:string) из десяти цифр: 0, 1, 2, 3, 4, 5, 6, 7, 8 и 9. В этой строке каждая цифра показывает, сколько раз число содержит определенную [степень](gloss:power) десятки. Например, когда записывают 213, имеют в виду следующее:

$$ 200+10+3 $$

Это же выражение можно записать таким образом, чтобы подчеркнуть степени 10:

$$ (2×10^2)+(1×10^1)+(3×10^0) $$

Хоть мы обычно используем систему, в основе которой лежит число 10, так же легко можно пользоваться системой, в основе которой лежит любое другое число. В основе двоичной системы счисления, например, лежит число 2. При этом для представления числа как суммы степеней двойки используется два символа: 1 и 2.  Например, число 213 будет представлено как 11010101, поскольку:

$$ \begin{aligned} 213 = &amp; \phantom{+}(1×2^7)+(1×2^6)+(0×2^5)\ &amp; +(1×2^4)+(0×2^3)+(1×2^2)\ &amp; +(0×2^1)+(1×2^0) \ \end{aligned} $$

Здесь мы выражаем числа суммой чисел, кратных 2, 4, 8, 16, 32 и т. д. вместо 10, 100, 1000 и т. д.

## Квантовые схемы

На предыдущей странице мы узнали, что в процессе вычислений над входными данными производятся определенные операции, и в результате получаются выходные данные. В случае с квантовыми компьютерами, о которых мы узнаем из этого учебника, такие данные всегда будут представлены в форме битов. Теперь, когда мы знаем, что такое биты, давайте посмотрим, какие действия мы можем с ними выполнять, чтобы превратить имеющиеся у нас входные данные в нужные нам выходные данные.

## Ваша первая квантовая схема

При работе с квантовой схемой обычно нужно выполнить три задачи: сначала закодировать входные значения, затем непосредственно выполнить вычисления и, наконец, извлечь выходные значения. Для нашей первой квантовой схемы мы сосредоточимся на последнем из этих этапов. Начнем с создания квантовой схемы с 3 кубитами и 3 выходами.

````python
from qiskit import QuantumCircuit
# Создаем квантовую схему с 3 кубитами и 3 классическими битами
# (чуть позже мы узнаем, для чего нужные классические биты)
qc = QuantumCircuit(3, 3)
qc.draw()  # возвращает изображение схемы 
````

В самом конце метод <code>qc.draw()</code> формирует изображение схемы. Jupyter Notebook вычисляет последнюю строку ячейки кода и отображает результат под ячейкой. Поскольку <code>qc.draw()</code> [возвращает](gloss:return) изображение, именно его мы видим под кодом. В нашей схеме еще нет гейтов, поэтому мы видим лишь несколько горизонтальных линий.

Далее нам нужен способ заставить наш квантовый компьютер измерить кубиты и записать результаты. Для этого мы добавляем в квантовую схему операцию «измерения». Мы можем сделать это с помощью инструкции `QuantumCircuit` `.measure()` .

````python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3, 3)
# измерить кубиты 0, 1 & 2, записав соответствующие значения в классические биты 0, 1 & 2
qc.measure([0,1,2], [0,1,2])
qc.draw()
````

Давайте посмотрим, что получится после запуска этой схемы. Для этого мы воспользуемся квантовым симулятором, который представляет собой обычный компьютер, который вычисляет, что бы сделал квантовый компьютер.

Считается, что классическим компьютерам сложно эмулировать квантовый компьютер, поскольку лучшие из доступных алгоритмов растут экспоненциально с количеством кубитов. Таким образом, эмулировать можно только для схемы с небольшим количеством кубитов (примерно до 30 кубитов), или же схемы конкретного вида, для которых можно использовать определенные приемы ускорения эмуляции. Тем не менее, эмуляторы — очень полезные инструменты для разработки небольших квантовых схем.

Давайте импортируем эмулятор Qiskit (он называется Aer) и создадим новый объект симулятора.

````python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator()  # создаем новый объект эмулятора
````

Для запуска эмуляции можно использовать метод <code>.run()</code> эмулятора. Он возвращает «задание» (job), которое содержит информацию об эксперименте, например, запущен ли эксперимент или завершен, о серверной части, на которой он выполнялся, и, самое главное, результат эксперимента.

Для получения результата выполнения задачи используется метод result, и самый распространенный способ представления результатов — использование словаря «счетчиков» (количества измерений).

````python
job = sim.run(qc)      # запускаем эксперимент
result = job.result()  # получаем результаты
result.get_counts()    # представляем результат в виде словаря "счетчиков" (counts)
````

Ключи в словаре counts — это битовые строки, а значения — это количество раз, которое данная битовая строка была измерена. В результатах работы квантовых компьютеров присутствует элемент случайности, поэтому обычно квантовая схема запускается несколько раз. В данном случае схема была запущена 1024 раза – именно столько раз в Qiskit схемы запускаются по умолчанию. Исходное состояние кубитов принято устанавливать в значение `0`, и, поскольку мы ничего не делаем с ними перед измерением, результаты всегда равны `0`.

### Кодирование входных значений

Теперь давайте посмотрим, как закодировать входные данные другой двоичной строкой. Для этого нам понадобится так называемый вентиль НЕ (NOT-гейт). Это самая базовая операция, можно выполнить на компьютере. Он просто инвертирует значение бита: 0 становится 1, а 1 становится 0. При работе с кубитами мы используем для этого вентиль, который называется *X-гейт* .

Здесь мы создаем новую квантовую схему, предназначенную для кодирования:

````python
# Создаем квантовую схему с 3 кубитами и 3 классическими битами:
qc = QuantumCircuit(3, 3)
qc.x([0,1])  # Применяем X-гейт к кубитам 0 & 1
qc.measure([0,1,2], [0,1,2])
qc.draw()    # возвращает изображение схемы
````

А для получения результатов запустим эмуляцию работы схемы:

````python
job = sim.run(qc)      # запускаем эксперимент
result = job.result()  # получаем результаты
result.get_counts()    # представляем результат в виде словаря "счетчиков" (counts)
````

### Сложение с помощью квантовых схем

Давайте создадим наш собственный полусумматор на основе квантовой схемы. Одна часть схемы будет кодировать ввод, вторая – алгоритм, третья – извлекать результат. Первую часть нужно будет изменять каждый раз, когда необходимо использовать новые входные значения, а остальные части остаются неизменными.

![изображение схемы сумматора](images/aoc/half-adder.svg)

Два бита, которые мы хотим сложить, закодированы в кубитах 0 и 1. Приведенный выше пример кодирует значение 1 для обоих кубитов, поэтому здесь вычисляется выражение 1+1. Результатом будет строка из двух битов, которую мы прочитаем из кубитов 2 и 3. Остается только заполнить собственно программу, которая располагается в пространстве посередине.

Пунктирные линии на изображении просто разделяют схему на разные части (хотя их также можно использовать и для более интересных вещей).

Базовые вычислительные операции называют логическими вентилями. Мы уже использовали вентиль НЕ (NOT-гейт), но для создания полусумматора его недостаточно. Мы использовали его только для ручного получения ответов. Поскольку мы хотим, чтобы компьютер выполнял за нас фактические вычисления, нам понадобятся более функциональные логические вентили.

Чтобы понять, что нам нужно, давайте еще раз взглянем на то, что должен делать наш полусумматор.

![операции, которые должен выполнять полусумматор](images/aoc/half_adder_reqs.svg)

Самый правый бит во всех этих четырех ответах полностью определяется тем, совпадают ли значения двух складываемых битов, или нет. Таким образом, для 0+0 и 1+1, где два бита равны, самый правый бит ответа равен 0. Для 0+1 и 1+0, где мы складываем биты с разными значениями, самый правый бит равен 1.

Для правильного вычисления этой части нам понадобится что-то, что может определить одинаковы ли значение двух битов, или нет. Обычно в цифровых вычислениях это называется XOR-гейт (вентиль исключающего логического ИЛИ).

В квантовых компьютерах функцию XOR-гейта (вентиль «исключающее ИЛИ») выполняет вентиль «управляемое НЕ» (гейт controlled-NOT). Поскольку это довольно длинное название, обычно его называют просто «CNOT». На принципиальных схемах он обозначается как на изображении ниже. Гейт работает с двумя кубитами. Один выполняет роль управляющего кубита (тот, что с маленькой точкой). Другой выступает в качестве целевого кубита (с большим кругом и крестом — похож на мишень).

![изображение CNOT-гейта](images/aoc/cnot_xor.svg)

В Qiskit можно использовать метод `.cx()` для добавления CNOT-гейта в схему. В качестве аргументов необходимо указать индексы двух кубитов, к которым он применяется. Вот пример:

````python
# Строим квантовую схему с 2 кубитами и 2 классическими битами
qc = QuantumCircuit(2, 2)
qc.x(0)
qc.cx(0,1)  # CNOT-гейт, управляемый кубитом 0 с целевым кубитом 1
qc.measure([0,1], [0,1])
display(qc.draw())     # изображение схемы

job = sim.run(qc)      # запускаем эксперимент
result = job.result()  # получаем результаты
# представляем результаты в виде словаря счетчиков ("counts")
print("Result: ", result.get_counts())
````

Нам не нужно, чтобы полусумматор перезаписывал один из наших входов. Вместо этого мы хотим сохранить результат в другой паре кубитов. Это можно сделать с помощью двух CNOT-гейтов, записав выходное значение в новый кубит, который, как мы знаем, будет в состоянии 0:

![Схема , использующая два CNOT-гейта для создания XOR-гейта](images/aoc/xor.svg)

Мы уже на полпути к полностью рабочему полусумматору. Нам известно, как вычислить крайний правый выходной бит – осталось решить, как вычислить левый. Если снова взглянуть на четыре возможные комбинации сумм, можно заметить, что только в одном случае левый бит равен 1, а не 0: 1+1=10. Это происходит только тогда, когда оба складываемых бита равны 1.

![операции, которые должен выполнять полусумматор](images/aoc/half_adder_reqs.svg)

Для вычисления этой части выходного значения мы можем бы просто заставить наш компьютер проверить, равны ли 1 оба входных значения. Если они равны — и только в этом случае — нам нужно применить NOT-гейт к кубиту 3. Только в этом случае он будет инвертирован, приняв нужное значение 1, в результате чего получится необходимое нам выходное значение.

Для этого нам нужен новый вентиль: как CNOT, но управляемый двумя кубитами вместо одного. Он будет выполнять операцию НЕ на целевом кубите только в том случае, если оба управляющих бита находятся в состоянии 1. Такой новый вентиль называется вентилем [Тоффоли](gloss:toffoli) (Toffoli gate). Для тех из вас, кто знаком с булевыми логическими вентилями, это по сути вентиль И (AND).

![квантовая схема полусумматора](images/aoc/ha_circuit.svg)

В Qiskit его можно добавиьт в схему с помощью метода `.ccx()` . Вот и все! Схема, которая может решить знаменитую математическую задачу 1+1.
